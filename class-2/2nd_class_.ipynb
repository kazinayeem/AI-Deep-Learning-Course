#ZIP Function

```python
ID = [1,2,3,4,5,6]
furniture = ["Table", "Chair", "Bed", "Rack", "Self", "Almirae"]
price = [1000, 100, 5000, 1200, 1400, 15000]
```

```python
for ID, furniture, price in zip(ID, furniture, price):
    print(f"The {ID} of {furniture} price is: {price}")
```
